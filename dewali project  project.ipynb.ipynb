```python
ax = sns.countplot(x = "Gender",data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(x = "Age Group",data = df,hue = "Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
var2 = df.groupby(["Age Group"], as_index=False)[["Amount", "Orders"]].sum().sort_values(by="Amount", ascending=False)
sns.barplot(x="Age Group", y="Orders", data=var2)
```

```python
df.columns
```

```python
var2 = df.groupby(["State"], as_index=False)[ "Orders"].sum().sort_values(by="Orders", ascending=False).head(15)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x="State", y="Orders", data=var2)
```

```python
var2 = df.groupby(["State"], as_index=False)[ "Amount"].sum().sort_values(by="Amount", ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x="State", y="Amount", data=var2)
```

```python
df.columns
```

```python
ax = sns.countplot(x = "Marital_Status",data = df)
sns.set(rc={'figure.figsize':(4,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
var2 = df.groupby(["Marital_Status","Gender"], as_index=False)[ "Amount"].sum().sort_values(by="Amount", ascending=False)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x="Marital_Status", y="Amount", data=var2 , hue = "Gender")
```

in the above diagram we clearly see that most of buyeers are married woman and its has high puschsing power


```python
var2 = df.groupby(["Occupation"], as_index=False)[ "Amount"].sum().sort_values(by="Amount", ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="Occupation", y="Amount", data=var2)
```

here we can clearly see that the most of the buyeers are itsector ,healthcare ,and aviation .

```python
ax = sns.countplot(x = "Occupation",data = df)
sns.set(rc={'figure.figsize':(4,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(x = "Product_Category",data = df)
sns.set(rc={'figure.figsize':(26,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
var2 = df.groupby(["Product_Category"], as_index=False)[ "Amount"].sum().sort_values(by="Amount", ascending=False).head(  10)
sns.set(rc={'figure.figsize':(26,7)})
sns.barplot(x="Product_Category", y="Amount", data=var2)
```

the most buying product in dewali is food ,clothin and appareland electronics gagets

```python
var2 = df.groupby(["Product_ID"], as_index=False)[ "Orders"].sum().sort_values(by="Orders", ascending=False).head(  10)
sns.set(rc={'figure.figsize':(26,7)})
sns.barplot(x="Product_ID", y="Orders", data=var2)
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby("Product_Category")["Orders"].sum().nlargest(10).sort_values(ascending = False).plot(kind="bar")
```
